## 🔬 Célula 4 — Algoritmo TF-IDF
Cada livro é representado por um vetor TF-IDF construído a partir de suas 3 características: **autor**, **tema** e **categoria**.

```python
documents = []

def tokenize(text):
    """Extrai tokens de um texto (aceita caracteres acentuados)."""
    return re.findall(r'\b[a-zA-ZÀ-ÿ]+\b', text.lower())

# Cria documento para cada livro combinando suas 3 características
for book in books:
    phrase = book["autor"] + " " + book["tema"] + " " + book["categoria"]
    documents.append(tokenize(phrase))

# Vocabulário único
word_set = list({word for doc in documents for word in doc})
index_dict = {word: i for i, word in enumerate(word_set)}
total_docs = len(documents)

# IDF
def compute_idf():
    word_count = {word: 0 for word in word_set}
    for word in word_set:
        for doc in documents:
            if word in doc:
                word_count[word] += 1
    return {word: np.log(total_docs / (word_count[word] + 1)) for word in word_set}

idf = compute_idf()

# TF
def termfreq(document, word):
    return document.count(word) / len(document)

# Vetor TF-IDF
def tf_idf(document):
    vec = np.zeros(len(word_set))
    for word in document:
        if word in index_dict:
            vec[index_dict[word]] = termfreq(document, word) * idf[word]
    return vec

# Similaridade de cosseno
def cosine_similarity(a, b):
    norm = np.linalg.norm(a) * np.linalg.norm(b)
    return np.dot(a, b) / norm if norm != 0 else 0

# Pré-calcula vetores de todos os livros
vectors = [tf_idf(doc) for doc in documents]

print(f'✅ TF-IDF calculado | Vocabulário: {len(word_set)} termos | Vetores: {len(vectors)}')
```

## 🎯 Célula 5 — Funções de Recomendação e Cadastro

```python
def take_top10_rating(idUser):
    """Retorna os 10 livros mais bem avaliados pelo usuário."""
    return sorted(range(len(ratings[idUser])), key=lambda i: ratings[idUser][i], reverse=True)[:10]

def recomend_books(idUser, top_n=5):
    """Gera recomendações baseadas no perfil do usuário via TF-IDF."""
    top10 = take_top10_rating(idUser)
    text = ""
    for i in top10:
        text += f" {books[i]['autor']} {books[i]['categoria']} {books[i]['tema']}"
    user_vec = tf_idf(tokenize(text))
    sims = [(i, cosine_similarity(user_vec, vec)) for i, vec in enumerate(vectors)]
    sims.sort(key=lambda x: x[1], reverse=True)
    return sims[:top_n]

def register_user(name):
    """Cadastra novo usuário e inicializa suas avaliações."""
    new_id = len(users)
    users.append({"id": new_id, "nome": name})
    ratings.append([0] * 50)
    return new_id

def rate_book(userId, bookId, rate):
    """Registra a avaliação de um livro (nota 1–5)."""
    if userId >= len(users):        return False, "Usuário não existe."
    if not (0 <= bookId < 50):      return False, "Livro não existe."
    if not (1 <= rate <= 5):        return False, "Nota inválida (1–5)."
    ratings[userId][bookId] = rate
    return True, "✅ Avaliação registrada!"

print('✅ Funções de recomendação e cadastro prontas!')
```

```python
# ── Estado da sessão ──────────────────────────────────────────
state = {"user_id": None}

# ── Estilos ───────────────────────────────────────────────────
style_title  = 'font-size:20px; font-weight:bold; color:#e94560;'
style_sub    = 'font-size:14px; color:#555; margin-bottom:10px;'
style_card   = ('background:#f7f7f7; border-left:4px solid #e94560;'
                'padding:10px 15px; margin:6px 0; border-radius:4px;')
style_badge  = ('background:#e94560; color:#fff; padding:2px 8px;'
                'border-radius:12px; font-size:12px;')
style_output = 'padding:10px; background:#fff; border-radius:8px;'

# ── Widgets globais ───────────────────────────────────────────
out_status = widgets.Output()
out_main   = widgets.Output()

def status_msg(msg, color='#2ecc71'):
    with out_status:
        clear_output(wait=True)
        display(HTML(f'<div style="background:{color};color:#fff;padding:10px;'
                     f'border-radius:6px;margin:6px 0">{msg}</div>'))

# ─────────────────────────────────────────────────────────────
# ABA 1 — Login / Cadastro
# ─────────────────────────────────────────────────────────────
user_options = [f"{u['id']} — {u['nome']}" for u in users]
sel_user     = widgets.Dropdown(options=user_options, description='Usuário:', layout=widgets.Layout(width='420px'))
btn_login    = widgets.Button(description='Entrar', button_style='success')
txt_name     = widgets.Text(placeholder='Nome completo', description='Nome:', layout=widgets.Layout(width='420px'))
btn_register = widgets.Button(description='Cadastrar', button_style='primary')

def on_login(_):
    uid = int(sel_user.value.split(' — ')[0])
    state['user_id'] = uid
    status_msg(f"✅ Bem-vindo(a), {users[uid]['nome']}!")

def on_register(_):
    name = txt_name.value.strip()
    if len(name) < 3:
        status_msg('⚠️ Digite um nome com ao menos 3 caracteres.', '#e74c3c')
        return
    nid = register_user(name)
    state['user_id'] = nid
    sel_user.options = [f"{u['id']} — {u['nome']}" for u in users]
    txt_name.value = ''
    status_msg(f"✅ Cadastrado! ID {nid} — {name}")

btn_login.on_click(on_login)
btn_register.on_click(on_register)

tab_login = widgets.VBox([
    HTML(f'<p style="{style_title}">👤 Login / Cadastro</p>'),
    HTML('<b>Usuário existente</b>'), sel_user, btn_login,
    HTML('<hr><b>Novo usuário</b>'), txt_name, btn_register
])

# ─────────────────────────────────────────────────────────────
# ABA 2 — Catálogo
# ─────────────────────────────────────────────────────────────
all_cats   = ['Todas'] + sorted({b['categoria'] for b in books})
sel_cat    = widgets.Dropdown(options=all_cats, description='Categoria:', layout=widgets.Layout(width='260px'))
out_cat    = widgets.Output()

def show_catalog(cat=None):
    with out_cat:
        clear_output(wait=True)
        filtered = [b for b in books if cat in (None, 'Todas', b['categoria'])]
        html = ''.join(
            f'<div style="{style_card}">[{b["id"]:02d}] <b>{b["nome"]}</b> — '
            f'{b["autor"]} &nbsp;<span style="{style_badge}">{b["categoria"]}</span>'
            f'<br><small style="color:#888">{b["tema"]}</small></div>'
            for b in filtered
        )
        display(HTML(html))

sel_cat.observe(lambda c: show_catalog(c['new']), names='value')
show_catalog()

tab_catalogo = widgets.VBox([
    HTML(f'<p style="{style_title}">📖 Catálogo de Livros</p>'),
    sel_cat, out_cat
])

# ─────────────────────────────────────────────────────────────
# ABA 3 — Avaliar
# ─────────────────────────────────────────────────────────────
book_options = [f"{b['id']:02d} — {b['nome']} ({b['autor']})" for b in books]
sel_book  = widgets.Dropdown(options=book_options, description='Livro:', layout=widgets.Layout(width='480px'))
sel_stars = widgets.RadioButtons(
    options=['⭐ 1', '⭐⭐ 2', '⭐⭐⭐ 3', '⭐⭐⭐⭐ 4', '⭐⭐⭐⭐⭐ 5'],
    description='Nota:', layout=widgets.Layout(width='360px')
)
btn_rate  = widgets.Button(description='Enviar Avaliação', button_style='warning')
out_hist  = widgets.Output()

def on_rate(_):
    if state['user_id'] is None:
        status_msg('⚠️ Faça login primeiro!', '#e74c3c')
        return
    bid  = int(sel_book.value.split(' — ')[0])
    nota = int(sel_stars.value.split()[1])
    ok, msg = rate_book(state['user_id'], bid, nota)
    status_msg(msg, '#2ecc71' if ok else '#e74c3c')
    show_history()

def show_history():
    with out_hist:
        clear_output(wait=True)
        if state['user_id'] is None:
            display(HTML('<i>Faça login para ver suas avaliações.</i>'))
            return
        rated = [(i, r) for i, r in enumerate(ratings[state['user_id']]) if r > 0]
        rated.sort(key=lambda x: x[1], reverse=True)
        if not rated:
            display(HTML('<i>Nenhuma avaliação ainda.</i>'))
            return
        html = ''.join(
            f'<div style="{style_card}"><b>{books[i]["nome"]}</b> — '
            f'{books[i]["autor"]} &nbsp;{'⭐'*r}</div>'
            for i, r in rated
        )
        display(HTML(html))

btn_rate.on_click(on_rate)

tab_avaliar = widgets.VBox([
    HTML(f'<p style="{style_title}">⭐ Avaliar Livros</p>'),
    sel_book, sel_stars, btn_rate,
    HTML('<hr><b>Suas avaliações</b>'), out_hist
])

# ─────────────────────────────────────────────────────────────
# ABA 4 — Recomendações
# ─────────────────────────────────────────────────────────────
spin_n    = widgets.BoundedIntText(value=5, min=1, max=20, description='Qtd:', layout=widgets.Layout(width='180px'))
btn_rec   = widgets.Button(description='🎯 Gerar Recomendações', button_style='info')
out_rec   = widgets.Output()

def on_recommend(_):
    if state['user_id'] is None:
        status_msg('⚠️ Faça login primeiro!', '#e74c3c')
        return
    recs = recomend_books(state['user_id'], spin_n.value)
    with out_rec:
        clear_output(wait=True)
        name = users[state['user_id']]['nome']
        display(HTML(f'<p style="{style_title}">Recomendações para {name}</p>'))
        html = ''.join(
            f'<div style="{style_card}">'
            f'<b>#{rank} {books[bid]["nome"]}</b> '
            f'<span style="{style_badge}">score {score:.4f}</span><br>'
            f'<small>{books[bid]["autor"]} | {books[bid]["categoria"]} | {books[bid]["tema"]}</small>'
            f'</div>'
            for rank, (bid, score) in enumerate(recs, 1)
        )
        display(HTML(html))

btn_rec.on_click(on_recommend)

tab_rec = widgets.VBox([
    HTML(f'<p style="{style_title}">🎯 Recomendações Personalizadas</p>'),
    spin_n, btn_rec, out_rec
])

# ─────────────────────────────────────────────────────────────
# ABA 5 — Sobre
# ─────────────────────────────────────────────────────────────
tab_sobre = widgets.VBox([HTML('''
<p style="font-size:20px;font-weight:bold;color:#e94560">ℹ️ Sobre o Projeto</p>
<b>Sistema de Recomendação de Livros Brasileiros</b><br>
Disciplina: Introdução à Inteligência Artificial — UnB 2026/1<br>
Professor: Díbio<br><br>
<b>Como funciona</b>
<ul>
  <li>Cada livro tem 3 características: autor, tema e categoria</li>
  <li>TF-IDF pondera a relevância de cada termo no corpus</li>
  <li>Similaridade de cosseno mede a proximidade entre perfis</li>
  <li>Recomendações são geradas a partir dos livros mais bem avaliados</li>
</ul>
<b>Dados</b>: 50 livros · 100 usuários · 5.000 avaliações<br><br>
<b>Integrantes</b><br>
• Leandro Coelho da Silva<br>
• Paulo Henrique Borges Martins<br>
• Valeria Alexandra Guevara Parra
''')])

# ─────────────────────────────────────────────────────────────
# MONTAGEM FINAL
# ─────────────────────────────────────────────────────────────
tabs = widgets.Tab(children=[tab_login, tab_catalogo, tab_avaliar, tab_rec, tab_sobre])
for i, title in enumerate(['👤 Login', '📖 Catálogo', '⭐ Avaliar', '🎯 Recomendações', 'ℹ️ Sobre']):
    tabs.set_title(i, title)

display(HTML('<h2 style="color:#e94560">📚 Sistema de Recomendação de Livros Brasileiros</h2>'))
display(out_status)
display(tabs)
```
