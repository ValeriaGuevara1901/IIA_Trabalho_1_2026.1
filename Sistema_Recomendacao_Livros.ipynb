## 4. Matriz de Utilidade

A **matriz de utilidade** representa as avaliações dos usuários para os livros. Cada célula `ratings[i][j]` contém a nota (0-5) que o usuário `i` deu ao livro `j`.

- **Dimensão**: 100 usuários × 50 livros = 5.000 avaliações
- **Escala**: 0 (não avaliado) a 5 (excelente)

Para fins deste projeto, a matriz foi gerada aleatoriamente com semente fixa para reprodutibilidade.

```python
# Geração da matriz de utilidade
random.seed(42)  # Semente para reprodutibilidade

ratings = []
for i in range(100):
    grades = []
    for j in range(50):
        grades.append(random.randint(0, 5))
    ratings.append(grades)

# Convertendo para numpy array para visualização
ratings_matrix = np.array(ratings)

print(f"Dimensões da matriz de utilidade: {ratings_matrix.shape}")
print(f"Total de avaliações: {ratings_matrix.size}")
print(f"\nPrimeiras 5 linhas e 10 colunas da matriz:")
print(ratings_matrix[:5, :10])
```

## 5. Pré-processamento: Tokenização

A **tokenização** é o processo de dividir o texto em unidades menores (tokens). Utilizamos expressões regulares para extrair apenas palavras, convertendo tudo para minúsculas.

```python
def tokenize(text):
    """
    Tokeniza o texto, extraindo apenas palavras (incluindo acentuadas).
    
    Parâmetros:
        text (str): Texto a ser tokenizado
    
    Retorna:
        list: Lista de tokens em minúsculas
    """
    return re.findall(r'\b[a-zA-ZÀ-ÿ]+\b', text.lower())

# Exemplo de tokenização
exemplo = "Machado de Assis - Romance psicológico"
tokens = tokenize(exemplo)
print(f"Texto original: '{exemplo}'")
print(f"Tokens: {tokens}")
```

## 6. Criação dos Documentos

Cada livro é representado por um **documento** que combina suas 3 características: autor, tema e categoria. Este documento será usado para calcular o TF-IDF.

```python
# Criação dos documentos para cada livro
documents = []

for book in books:
    # Combina as 3 características do livro
    phrase = f"{book['autor']} {book['tema']} {book['categoria']}"
    tokens = tokenize(phrase)
    documents.append(tokens)

# Exemplo
print(f"Livro: {books[0]['nome']}")
print(f"Documento: {documents[0]}")
```

## 7. Construção do Vocabulário

O **vocabulário** é o conjunto de todas as palavras únicas presentes nos documentos. Cada palavra recebe um índice único para ser representada em vetores.

```python
# Construção do vocabulário único
word_set = set()
for doc in documents:
    word_set.update(doc)

word_set = list(word_set)
index_dict = {word: i for i, word in enumerate(word_set)}
total_docs = len(documents)

print(f"Tamanho do vocabulário: {len(word_set)} palavras")
print(f"Total de documentos: {total_docs}")
print(f"\nExemplo de mapeamento: 'romance' -> índice {index_dict.get('romance', 'não encontrado')}")
```

## 8. Cálculo do IDF (Inverse Document Frequency)

O **IDF** mede a importância de uma palavra no corpus. Palavras que aparecem em muitos documentos têm IDF baixo, enquanto palavras raras têm IDF alto.

**Fórmula:** $IDF(t) = \log\left(\frac{N}{df(t) + 1}\right)$

Onde:
- $N$ = número total de documentos
- $df(t)$ = número de documentos que contêm o termo $t$

```python
def compute_idf():
    """
    Calcula o IDF (Inverse Document Frequency) para cada palavra do vocabulário.
    
    Retorna:
        dict: Dicionário com o valor IDF de cada palavra
    """
    # Conta quantos documentos contêm cada palavra
    word_count = {word: 0 for word in word_set}
    
    for word in word_set:
        for doc in documents:
            if word in doc:
                word_count[word] += 1
    
    # Calcula o IDF
    idf = {}
    for word in word_set:
        idf[word] = np.log(total_docs / (word_count[word] + 1))
    
    return idf

idf = compute_idf()

# Mostra algumas palavras com maiores e menores IDF
sorted_idf = sorted(idf.items(), key=lambda x: x[1], reverse=True)
print("Palavras com maior IDF (mais raras):")
for word, value in sorted_idf[:5]:
    print(f"  {word}: {value:.4f}")

print("\nPalavras com menor IDF (mais comuns):")
for word, value in sorted_idf[-5:]:
    print(f"  {word}: {value:.4f}")
```

## 9. Cálculo do TF (Term Frequency)

O **TF** mede a frequência de uma palavra em um documento específico.

**Fórmula:** $TF(t, d) = \frac{f(t, d)}{|d|}$

Onde:
- $f(t, d)$ = número de vezes que o termo $t$ aparece no documento $d$
- $|d|$ = número total de termos no documento

```python
def termfreq(document, word):
    """
    Calcula a frequência do termo (TF) em um documento.
    
    Parâmetros:
        document (list): Lista de tokens do documento
        word (str): Palavra a ser calculada
    
    Retorna:
        float: Frequência normalizada do termo
    """
    return document.count(word) / len(document)

# Exemplo
doc_exemplo = documents[0]
print(f"Documento: {doc_exemplo}")
print(f"TF de 'assis': {termfreq(doc_exemplo, 'assis'):.4f}")
print(f"TF de 'romance': {termfreq(doc_exemplo, 'romance'):.4f}")
```

## 10. Cálculo do TF-IDF

O **TF-IDF** combina TF e IDF para dar peso às palavras baseado em sua importância local (no documento) e global (no corpus).

**Fórmula:** $TF\text{-}IDF(t, d) = TF(t, d) \times IDF(t)$

```python
def tf_idf(document):
    """
    Calcula o vetor TF-IDF para um documento.
    
    Parâmetros:
        document (list): Lista de tokens do documento
    
    Retorna:
        numpy.ndarray: Vetor TF-IDF do documento
    """
    vec = np.zeros(len(word_set))
    
    for word in document:
        if word in index_dict:
            tf = termfreq(document, word)
            vec[index_dict[word]] = tf * idf[word]
    
    return vec

# Calcula TF-IDF para todos os livros
vectors = [tf_idf(doc) for doc in documents]

print(f"Quantidade de vetores TF-IDF: {len(vectors)}")
print(f"Dimensão de cada vetor: {len(vectors[0])}")
```

## 11. Similaridade de Cosseno

A **similaridade de cosseno** mede o ângulo entre dois vetores, independente de sua magnitude. Valores próximos de 1 indicam alta similaridade.

**Fórmula:** $\cos(\theta) = \frac{\vec{a} \cdot \vec{b}}{||\vec{a}|| \times ||\vec{b}||}$

```python
def cosine_similarity(a, b):
    """
    Calcula a similaridade de cosseno entre dois vetores.
    
    Parâmetros:
        a, b (numpy.ndarray): Vetores a serem comparados
    
    Retorna:
        float: Similaridade entre 0 e 1
    """
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    
    if norm_a == 0 or norm_b == 0:
        return 0
    
    return np.dot(a, b) / (norm_a * norm_b)

# Exemplo: similaridade entre dois livros
sim = cosine_similarity(vectors[0], vectors[1])  # Dom Casmurro vs Memórias Póstumas
print(f"Similaridade entre '{books[0]['nome']}' e '{books[1]['nome']}': {sim:.4f}")

sim2 = cosine_similarity(vectors[0], vectors[49])  # Dom Casmurro vs O Vilarejo
print(f"Similaridade entre '{books[0]['nome']}' e '{books[49]['nome']}': {sim2:.4f}")
```

## 12. Sistema de Recomendação

O sistema de recomendação funciona em três etapas:

1. **Identificar preferências**: Encontra os 10 livros mais bem avaliados pelo usuário
2. **Criar perfil**: Combina as características desses livros em um vetor TF-IDF
3. **Calcular similaridade**: Compara o perfil com todos os livros e recomenda os mais similares

```python
def take_top10_rating(idUser):
    """
    Retorna os IDs dos 10 livros com maiores notas do usuário.
    
    Parâmetros:
        idUser (int): ID do usuário
    
    Retorna:
        list: IDs dos 10 livros mais bem avaliados
    """
    return sorted(range(len(ratings[idUser])), 
                  key=lambda i: ratings[idUser][i], 
                  reverse=True)[:10]

def get_recommendations(userId, top_n=5):
    """
    Gera recomendações personalizadas para um usuário.
    
    Parâmetros:
        userId (int): ID do usuário
        top_n (int): Número de recomendações
    
    Retorna:
        list: Lista de tuplas (book_id, score)
    """
    # Pega os livros favoritos do usuário
    top10_books = take_top10_rating(userId)
    
    # Cria um texto com as características dos livros favoritos
    text = ""
    for i in top10_books:
        text += f" {books[i]['autor']} {books[i]['categoria']} {books[i]['tema']}"
    
    # Calcula o vetor TF-IDF do perfil do usuário
    tokens = tokenize(text)
    user_vec = tf_idf(tokens)
    
    # Calcula similaridade com todos os livros
    similaridades = []
    for i, vec in enumerate(vectors):
        sim = cosine_similarity(user_vec, vec)
        similaridades.append((i, sim))
    
    # Ordena por similaridade e retorna os top N
    similaridades.sort(key=lambda x: x[1], reverse=True)
    return similaridades[:top_n]

print("Funções de recomendação definidas com sucesso!")
```

## 13. Demonstração do Sistema

```python
# Demonstração para o usuário 0
user_id = 0
user_name = users[user_id]['nome']

print(f"📚 Recomendações para: {user_name}")
print("=" * 60)

# Mostra os livros favoritos do usuário
print(f"\n📖 Livros mais bem avaliados por {user_name}:")
top10 = take_top10_rating(user_id)
for i, book_id in enumerate(top10[:5], 1):
    book = books[book_id]
    nota = ratings[user_id][book_id]
    print(f"   {i}. {book['nome']} ({book['autor']}) - Nota: {nota}")

# Mostra as recomendações
print(f"\n🎯 Livros recomendados:")
recommendations = get_recommendations(user_id, top_n=5)
for rank, (book_id, score) in enumerate(recommendations, 1):
    book = books[book_id]
    print(f"   {rank}. {book['nome']}")
    print(f"      Autor: {book['autor']}")
    print(f"      Categoria: {book['categoria']}")
    print(f"      Score: {score:.4f}")
    print()
```

## 14. Funções de Cadastro e Avaliação

```python
def register_user(name):
    """
    Registra um novo usuário no sistema.
    
    Parâmetros:
        name (str): Nome do novo usuário
    
    Retorna:
        int: ID do novo usuário
    """
    new_id = len(users)
    users.append({"id": new_id, "nome": name})
    ratings.append([0] * 50)  # Inicializa avaliações com 0
    return new_id

def rate_book(userId, bookId, rate):
    """
    Registra a avaliação de um livro por um usuário.
    
    Parâmetros:
        userId (int): ID do usuário
        bookId (int): ID do livro
        rate (int): Nota de 1 a 5
    
    Retorna:
        tuple: (sucesso, mensagem)
    """
    if userId >= len(users):
        return False, "Usuário não existe."
    if bookId < 0 or bookId >= 50:
        return False, "Livro não existe."
    if rate < 1 or rate > 5:
        return False, "Nota inválida (deve ser entre 1 e 5)."
    
    ratings[userId][bookId] = rate
    return True, "Avaliação registrada com sucesso!"

# Exemplo: Cadastrar novo usuário
novo_id = register_user("Maria Silva")
print(f"Novo usuário cadastrado com ID: {novo_id}")

# Avaliar alguns livros
rate_book(novo_id, 0, 5)  # Dom Casmurro - 5 estrelas
rate_book(novo_id, 1, 4)  # Memórias Póstumas - 4 estrelas
rate_book(novo_id, 10, 5)  # Quincas Borba - 5 estrelas
print("Avaliações registradas!")

# Gerar recomendações
print(f"\n🎯 Recomendações para {users[novo_id]['nome']}:")
recs = get_recommendations(novo_id, 3)
for rank, (book_id, score) in enumerate(recs, 1):
    print(f"   {rank}. {books[book_id]['nome']} (Score: {score:.4f})")
```
